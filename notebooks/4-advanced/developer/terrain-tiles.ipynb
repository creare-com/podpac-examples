## TerrainTiles

```python
from podpac.datalib.terraintiles import TerrainTiles
from podpac import Coordinates, clinspace
import numpy as np
%matplotlib inline
```

```python
node = TerrainTiles(zoom=2)
```

```python
c = Coordinates([clinspace(43, 0, 1000), clinspace(-73, 0, 1000)], dims=['lat', 'lon'])
data = node.eval(c)
```

```python
data
```

```python
data.plot()
```
